```python
from datetime import datetime
import backtrader as bt
import matplotlib.pyplot as plt
```

```python
class Strat(bt.Strategy):
    def __init__(self):
        self.UPRO = self.getdatabyname('UPRO')
        self.TMF = self.getdatabyname('TMF')
        self.counter = 0
        
    def next(self):
        if  self.counter % 20 == 0:
            self.order_target_percent(self.UPRO, target=.5)
            self.order_target_percent(self.TMF, target=.5)
        self.counter += 1


start = datetime(1999,1,1)
end = datetime(2019,1,1)

cerebro = bt.Cerebro()

upro = bt.feeds.YahooFinanceData(dataname='UPRO',
                                 fromdate=start,
                                 todate=end)

tmf = bt.feeds.YahooFinanceData(dataname='TMF',
                                 fromdate=start,
                                 todate=end)

cerebro.adddata(upro)
cerebro.adddata(tmf)

cerebro.addstrategy(Strat)
cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio') 

results = cerebro.run()
cerebro.plot()
```

```python
import pyfolio as pf
```

```python

strat = results[0]
pyfoliozer = strat.analyzers.getbyname('pyfolio')
returns, positions, transactions, gross_lev = pyfoliozer.get_pf_items()
pf.create_simple_tear_sheet(returns, positions=positions, transactions=transactions,)
```
